```python
import libs.local
import pandas as pd
import seaborn as sns
```

```python
experiment = 'ssp585'
key = 'Ensemble mean'
seasons = ['DJF', 'MAM', 'JJA', 'SON']
region = 'All'

kwargs = {
    'experiment': experiment,
    'region': region,
    'suffix': '_delta_1980-2010'
}

siconc = libs.local.get_ensemble_series('siconc', **kwargs)[key] / 1e6
simass = libs.local.get_ensemble_series('simass', **kwargs)[key]
sithick = libs.local.get_ensemble_series('sithick', **kwargs)[key]
sisnthick = libs.local.get_ensemble_series('sisnthick', **kwargs)[key]
tas = libs.local.get_ensemble_series('tas', **kwargs)[key]
evspsbl = libs.local.get_ensemble_series('evspsbl', **kwargs)[key]
pr = libs.local.get_ensemble_series('pr', **kwargs)[key]
prra = libs.local.get_ensemble_series('prra', **kwargs)[key]
prsn = libs.local.get_ensemble_series('prsn', **kwargs)[key]
```

```python
data = pd.DataFrame({
    'siconc': siconc.data,
    'simass': simass.data,
    'sithick': sithick.data,
    'sisnthick': sisnthick.data,
    'tas': tas.data,
    'prra': prra.data,
    'prsn': prsn.data,
    'evspsbl': evspsbl.data
})
ax = sns.heatmap(data.corr(), cmap='RdBu', annot=True)
ax.set_title('Anomaly correlation');
```

```python
period = 'MAM'

data = pd.DataFrame({
    'siconc': siconc.where(siconc.time['time.season'] == period).data,
    'simass': simass.where(simass.time['time.season'] == period).data,
    'sithick': sithick.where(sithick.time['time.season'] == period).data,
    'sisnthick': sisnthick.where(sisnthick.time['time.season'] == period).data,
    'tas': tas.where(tas.time['time.season'] == period).data,
    'prra': prra.where(prra.time['time.season'] == period).data,
    'prsn': prsn.where(prsn.time['time.season'] == period).data,
    'evspsbl': evspsbl.where(evspsbl.time['time.season'] == period).data
})
ax = sns.heatmap(data.corr(), cmap='RdBu', annot=True)
ax.set_title(f'{period} anomaly correlation');
```
