```python
import libs.analysis
import libs.ensemble
import libs.plot
import matplotlib

matplotlib.rcParams.update({ 'font.size': 18 })
```

## Data

```python
# ----- SPECIFIC SETTINGS -----
component = 'SImon'
experiment = 'ssp585'
variable_id = 'simpconc'

spatial_text = 'sea-ice melt pond concentration'
spatial_units = '%'

colormesh_kwargs = {
    'extend': 'neither',
    'levels': 21,
    #'shading': 'auto', # NB change to 'gouraud' if Jaspy
    'vmin': 0,
    'vmax': 100,
    'x': 'longitude', 
    'y': 'latitude'
}
```

```python
ensemble, weight = libs.ensemble.get_and_preprocess(
    component, 
    experiment, 
    variable_id
)

# Generate slices
ensemble_time_slices = libs.analysis.generate_slices(ensemble)
```

## Plots

### April

```python
import warnings
warnings.filterwarnings('ignore')

libs.plot.calendar_division_spatial(
    ensemble_time_slices, 
    colormesh_kwargs=colormesh_kwargs,
    text=spatial_text,
    time='APR',
    title='{label} {time} ssp585 {text} 60-90N ({units})',
    units=spatial_units
)
```
